```python
import matplotlib.pyplot as plt
import numpy as np

A=np.array([230,560,780,127,128])
B=np.array([200,160,270,128,400])
a=np.diff(A)
b=np.diff(B)
print(a)
print(b)
year=np.array([1,2,3,4])
plt.plot(year,a,color='red',linestyle=':',label='A')
plt.plot(year,b,color='black',linestyle='-',label='B')
plt.legend()
plt.title("Profits for A and B")
plt.show()
```
